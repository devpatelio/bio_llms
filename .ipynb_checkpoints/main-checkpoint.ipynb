```python
import json
import requests
import pandas as pd
import matplotlib.pyplot as plt

search_term = 'sc_RNA_B-cells'
api = 'https://www.proteinatlas.org/api/search_download.php?search={search_term}&format=json&columns=g,gs&compress=no'
request = requests.get(api)

assert 200 == request.status_code ## ensure its 200
path = 'rna_immune_cell_monaco.tsv'

data = pd.read_csv(path, sep='\t')
data.head()

fig, ax = plt.subplots()
data['Immune cell'].value_counts().plot(ax=ax, kind='bar')

cats = ['Switched memory B-cell', 'Non-switched memory B-cell', 'Exhausted memory B-cell', 'naive B-cell']
df = data.loc[data['Immune cell'].isin(cats)]
```

```python
META = 'https://ftp.uniprot.org/pub/databases/uniprot/knowledgebase/complete/docs/cdlist.txt'
df = df.sort_values(by='Immune cell')
df.head()
```
